```python
#CODE FOR FILLING ALL THE MISSING VALUES IN THE COLUMNS IN PYTHON BY USING MEAN
for i in range(len(data.columns)):
    for i in data.columns:
        if data[i].isnull().sum()!=0:
            data[i].fillna(data[i].mean(), inplace=True)
        elif data[i].isnull().sum()==0:
            continue
        
```

```python
#CODE FOR FILLING ALL THE MISSING VALUES IN THE COLUMNS IN PYTHON BY USING MODE
for i in range(len(data.columns)):
    for i in data.columns:
        if data[i].isnull().sum()!=0:
            data[i].fillna(data[i].mode(), inplace=True)
        elif data[i].isnull().sum()==0:
            continue
        
```

```python
#CODE FOR FILLING ALL THE MISSING VALUES IN THE COLUMNS IN PYTHON BY USING MEDIAN
for i in range(len(data.columns)):
    for i in data.columns:
        if data[i].isnull().sum()!=0:
            data[i].fillna(data[i].median(), inplace=True)
        elif data[i].isnull().sum()==0:
            continue
        
```

```python
#LINE PLOT
plt.plot(input,output,color="r",marker="o")
plt.title('CARBON DI OXIDE PRODUCED BY ALL SECTORS OF KENYA IN 2017')
plt.xlabel('SECTORS')
plt.ylabel('2017')
plt.rcParams["figure.figsize"] = (40,6)
plt.show()
```

```python
output1=kenya['2018']
#LINE PLOT
plt.plot(input,output1,color="g",marker="o")
plt.title('CARBON DI OXIDE PRODUCED BY ALL SECTORS OF KENYA IN 2018')
plt.rcParams["figure.figsize"] = (40,6)
plt.xlabel('SECTORS')
plt.ylabel('2018')
plt.show()
```

```python
output2=kenya['2016']
#LINE PLOT
plt.plot(input,output2,color="b",marker="o")
plt.title('CARBON DI OXIDE PRODUCED BY ALL SECTORS OF KENYA IN 2016')
plt.xlabel('SECTORS')
plt.ylabel('2016')
plt.show()
```

# COMPARISON OF CARBON DI OXIDE PRODUCED BY ALL SECTORS OF KENYA IN 2016 AND 2017

```python
#LINE PLOT
plt.plot(input,output,"mo--",linewidth=3,markersize=10,label="2017")
plt.plot(input,output2,"yo:",linewidth=3,markersize=10,label="2016")
plt.title('CARBON DI OXIDE PRODUCED BY ALL SECTORS OF KENYA IN 2016 and 2017',fontsize="20")
plt.rcParams["figure.figsize"] = (100,6)
plt.xlabel('SECTORS',fontsize="20")
plt.ylabel('YEAR',fontsize="20")
plt.show()
```
