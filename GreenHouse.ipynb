```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
greenhouse_df = pd.read_csv('historical_emissions/historical_emissions.csv')
```

```python
greenhouse_df[greenhouse_df.Country=="Kenya"]
```

## Extract Kenya Data

```python
kenya_df = greenhouse_df[greenhouse_df.Country=="Kenya"]
```

```python
kenya_df
```

## Reorganise Data

Create initial dataframe

```python
kenya_reorg_df = kenya_df[['Country','Data source', 'Sector', 'Gas', 'Unit', '1990']].copy()
```

```python
kenya_reorg_df
```

Rename the year column

```python
kenya_reorg_df.rename(columns = {'1990':'Reading'}, inplace = True)
```

```python
kenya_reorg_df
```

Add new column to contain the year

```python
kenya_reorg_df['DataYear'] = 1990
```

```python
kenya_reorg_df
```

Now loop through the remaining years and append to the new initial dataframe.

```python
for curr_year in range(1991, 2019):
    str_year = str(curr_year)
    copied_df = kenya_df[['Country','Data source', 'Sector', 'Gas', 'Unit', str_year]].copy()
    copied_df.rename(columns = {str_year:'Reading'}, inplace = True)
    copied_df['DataYear'] = curr_year
    #kenya_reorg_df = kenya_reorg_df.append(copied_df, ignore_index = True) #deprecated Use concat (below) instead
    kenya_reorg_df = pd.concat([kenya_reorg_df, copied_df])
    print(str_year, end= "...")
print("Done")
```

```python
kenya_reorg_df.head(42)
```

```python
kenya_reorg_df.Sector.unique()
```

```python
sector = "Energy"#Enter desired sector from list above
```

```python
plt.plot(kenya_reorg_df[kenya_reorg_df.Sector==sector]['DataYear'], kenya_reorg_df[kenya_reorg_df.Sector==sector]['Reading'])  # Plot the chart
plt.title("Linear Graph for " + sector + " Sector")
plt.show()  # display
```
