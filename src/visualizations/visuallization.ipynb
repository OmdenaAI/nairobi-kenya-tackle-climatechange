```python
XCO2 = dataset.variables['XCO2']
lats = dataset.variables['lat']
lons = dataset.variables['lon']
time = dataset.variables['time']
```

# ploting visualization XCO2 dataon kenya

```python
fig = plt.gcf()
fig.set_size_inches(9,12)

mp = Basemap(projection='merc', 
            llcrnrlat=-4.866317,  llcrnrlon= 33.663704,  #latitude and longitude of kenya
                  urcrnrlat= 5.292284 , urcrnrlon= 42.219556,
                  resolution='i')


lon, lat = np.meshgrid(lons, lats)
x, y = mp(lon, lat)

months = np.arange(0,84)

for i in months:
    
    c_plot = mp.pcolor(x, y, np.squeeze(XCO2[i,:,:]), cmap = 'jet')
    mp.drawlsmask()
    mp.drawstates(color='black' , linewidth = 1)

    mp.drawcoastlines(color='white', linewidth= 1)

    mp.drawcountries(color='black' , linewidth = 1)

    cbar = plt.colorbar( orientation = 'vertical', format = '%.6f',
                        extend = 'both', shrink = 0.45, aspect = 10, pad= 0.04)


    cbar.set_label(label = 'XCO2 (parts per million)' , size = 15)

    cbar.ax.tick_params(labelsize = 15)
    
    months= i + 1

    max_xco2 = 0.000425
    min_xco2 = 0.000390

    cbar.set_ticks(np.linspace(min_xco2,max_xco2, 5))

    plt.clim(min_xco2,max_xco2)

    plt.title('atmospheric carbon dioxide: month'+ str(months) , fontsize= 19)
    plt.savefig(r'D:/omdena project/visual'+'//'+str(months)+'.jpg')

    plt.clf()
```
