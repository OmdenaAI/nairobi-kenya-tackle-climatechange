# Linear Regression Model

```python
#TO PRE-PROCESS DATA\n",
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
#TO READ CSV FILE
data = pd.read_csv("final_data_combined.csv")
```

```python
data.head(25)
```

```python
data.rename(columns = {'XCO2 (mol CO2/mol dry)':'xco2_level'}, inplace = True)
```

```python
oil_df = data[(data['primary_fuel'] == 'Oil')]
oil_df
```

```python
#Create Linear Regression object
reg = linear_model.LinearRegression()
```

```python
reg.fit(oil_df[['commissioning_year', 'capacity_mw', 'frp' ]], oil_df.xco2_level) #Train the model using the dataset and specify the target variable
```

```python
#Look at the co-efficient
reg.coef_
```

```python
reg.intercept_
```

```python
#Predict
reg.predict([[2022, 150, 40]])
```

Work out the calculation using the coefficients and the intercept above.

```python
#coefs -2.67243027e-08, -3.97761072e-09, -4.82460259e-08
-2.67243027e-08*2022+-3.97761072e-09*150+-4.82460259e-08*40+0.0004614574872252408
```
